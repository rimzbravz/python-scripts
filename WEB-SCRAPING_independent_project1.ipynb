```python
import urllib
import bs4

#urllib2 is already incorporated in urllib
ASUS = "http://www.studyabroad.com/pages/sitecontent/telephonecodes.aspx"

#Query the Website store in variable Page

page = urllib.request.urlopen(ASUS)

#import the Beautiful soup functions to parse the data returned from the website

from bs4 import BeautifulSoup

soup = BeautifulSoup(page)

#let's look at nested structured of HTML page
print (soup.prettify())


```

```python
#Lets find the table where we want to return
#but first, return all tables

all_tables = soup.find_all('div') 
all_tables
```

```python
right_table = soup.find_all('a', {"class": 'price'}) #main box subdivided into <img alt=...
print(len(right_table))
right_table

```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table[0].findAll("tr"): #mag-iiterate siya sa lahat ng tr!
    cells = row.findAll('td')
    states=row.findAll('th') #To store second column data
    if len(cells)==6: #Only extract table body not heading
        A.append(cells[0].find(text=True)) #to access value of each element use find(text=True), remember ung string lang kukunin 
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
```

```python
col_headings = right_table[0].findAll("th", {"scope" : "col"}) #return the column Headings
col_headings
```

```python
#use pandas to convert list to dat frame, kumbaga sa R, x <- data.frame()
import pandas as pd

df = pd.DataFrame(A, columns=['Number']) #define how the df behaves
df['State/UT'] = B #apply to other columns using ?function df
df['Admin_Capital'] = C
df['Legislative_Capital'] = D
df['Judiciary_Capital'] = E
df['Capital_Year_Est'] = F
df['Former_Capital'] = G

df
```
