```python
#The importance of preprocessing in data science and the machine learning pipeline II: centering, scaling and logistic regression

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('ggplot')

from sklearn import datasets
from sklearn import linear_model

import numpy as np

#Load Data
boston = datasets.load_boston()

yb = boston.target.reshape(-1,1)
Xb = boston['data'][:,5].reshape(-1,1) #reshape transforms into a list

plt.scatter(Xb,yb)
plt.ylabel('Value of House /1000($)')
plt.xlabel('number of rooms')

#create linear regression object

regr = linear_model.LinearRegression()
regr.fit(Xb, yb)

#plot outputs

plt.scatter(Xb, yb, color = 'red')
plt.plot(Xb, regr.predict(Xb), color = 'black', linewidth = 2)
plt.show()
```
